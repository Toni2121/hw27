```python
import pandas as pd
import numpy as np

mobile = pd.read_csv(r"mobile.csv")

# -----------------------------------------------------------------------------------------------
# 1.
mobile['First Name'] = mobile['Customer Name'].apply(lambda value: str(value).split(" ")[0])
print()
# -----------------------------------------------------------------------------------------------
# 2.
mobile['Phone Company'] = mobile['Phone Number'].apply(
    lambda value: "Pelephone" if str(value)[:3] == "050" else
                  "Golan" if str(value)[:3] == "051" else
                  "Cellcom" if str(value)[:3] == "052" else
                  "Motorolla" if str(value)[:3] == "053" else
                  "Hot"
)
print()
# -----------------------------------------------------------------------------------------------
# 3.
# a.
print(mobile['Phone Company'].value_counts())
print()
# b.
print(f"number of companies: \n{list(mobile['Phone Company'].unique())}")
print()
# c.
print(f"number of companies: \n{mobile['Phone Company'].nunique()}")
print()
# -----------------------------------------------------------------------------------------------
# 4.
# a.
mobile['User Type'] = mobile.apply(lambda Type: 'Talker' if Type['Talked Minutes'] > Type['Internet Minutes'] else "Surfer", axis=1)
print()
# b.
mobile['User Type v'] = np.vectorize(lambda Talked_Minutes, Internet_Minutes: "Talker" if Talked_Minutes > Internet_Minutes else "Surfer")(mobile['Talked Minutes'], mobile['Internet Minutes'])
print()
# -----------------------------------------------------------------------------------------------
# 5.
sort_by_time_talked = mobile.sort_values(['Talked Minutes'], ascending=True)
print(sort_by_time_talked)
print()
# -----------------------------------------------------------------------------------------------
# 6.
sort_by_time_talked_and_surfed = mobile.sort_values(['Talked Minutes', 'Internet Minutes'], ascending=[True, True])
print(sort_by_time_talked_and_surfed)
print()
# -----------------------------------------------------------------------------------------------
# 7.
most_talked_customers = mobile.nlargest(5, 'Talked Minutes')
print(f"Most talked customers: \n{most_talked_customers}")
print()
# -----------------------------------------------------------------------------------------------
# 8.
least_surfing_customers = mobile.nsmallest(3, 'Internet Minutes')
print(f"least surfing customers: \n{least_surfing_customers}")
print()
# -----------------------------------------------------------------------------------------------
# 9.
# a.
duplicate_rows = mobile[mobile.duplicated()]
print(duplicate_rows)
print()
# b.
print(mobile.drop_duplicates())
print()
# -----------------------------------------------------------------------------------------------
# 10.
print(mobile["Customer Name"]["Phone Number"].sample(3))
mobile
```
